```python
#Normalizer for the dependent variable like temp
from sklearn.preprocessing import Normalizer
import pandas
import numpy
data = pandas.read_csv("airquality.csv")
mean_Solar=data['Solar.R'].mean()
mean_Solar=data['Ozone'].mean()
data['Solar.R'].fillna(mean_Solar,inplace=True)
data['Ozone'].fillna(mean_Ozone,inplace=True)
array = data.values
X = array[:,0:5]
Y = array[:,3]
scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
numpy.set_printoptions(precision=3)
print(normalizedX[0:5,:])
```

```python
#min max scalar
from sklearn.preprocessing import MinMaxScaler
import pandas
import numpy
array = data.values
X = array[:,0:5]
Y = array[:,3]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
numpy.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

```python
#standard scalar
from sklearn.preprocessing import StandardScaler
import pandas
import numpy
array = data.values
X = array[:,0:5]
Y = array[:,3]
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
numpy.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

```python
#Algorithms (models)
#classifiers
#decison classifier
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
array = data.values
seed = 7
models = []
#models.append(('LR', LogisticRegression()))
#models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f" % (name, cv_results.mean()*100)
    print(msg)
```

```python
# kneighbourclassifier
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
array = data.values
seed = 7
models = []
models.append(('KNN', KNeighborsClassifier()))
#models.append(('CART', DecisionTreeClassifier()))
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f" % (name, cv_results.mean()*100)
    print(msg)
```

```python
# Random Forest Classification
import pandas
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
data = pandas.read_csv("airquality.csv")
mean_Solar=data['Solar.R'].mean()
mean_Solar=data['Ozone'].mean()
data['Solar.R'].fillna(mean_Solar,inplace=True)
data['Ozone'].fillna(mean_Ozone,inplace=True)
array = data.values
X = array[:,0:5]
Y = array[:,3]
seed = 7
num_trees = 100
max_features = 3
kfold = model_selection.KFold(n_splits=10, random_state=seed)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean()*100)
```

```python
# Regressions
#decision tree regression
import pandas
from sklearn import model_selection
from sklearn.tree import DecisionTreeRegressor
data = pandas.read_csv("airquality.csv")
data['Solar.R'].fillna(mean_Solar,inplace=True)
data['Ozone'].fillna(mean_Ozone,inplace=True)
array = data.values
X = array[:,0:5]
Y = array[:,3]
seed = 7
kfold = model_selection.KFold(n_splits=10, random_state=seed)
model = DecisionTreeRegressor()
scoring = 'accuracy'
results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print(results.mean()*100)
```

```python
# logistic regression
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
array = data.values
seed = 7
models = []
models.append(('LR', LogisticRegression()))
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f" % (name, cv_results.mean()*100)
    print(msg)
```

```python
# Evaluate using a train and a test set
from sklearn.preprocessing import StandardScaler
import pandas
import numpy
array = data.values
X = array[:,0:5]
Y = array[:,3]
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
numpy.set_printoptions(precision=3)
print(rescaledX[0:5,:])
import pandas
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
array = data.values
X = array[:,0:5]
Y = array[:,3]
test_size = 0.30
seed = 5
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=test_size, random_state=seed)
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print("Accuracy: %.3f%%" % (result*100.0))
```
